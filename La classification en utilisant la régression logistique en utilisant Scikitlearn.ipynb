```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

```python
dataset = pd.read_csv("titanic.csv")
dataset.head()
```

```python
X = dataset[['Fare', 'Age','Pclass','Siblings/Spouses Aboard','Parents/Children Aboard']].values
y = dataset[['Survived']].values
x1 = dataset[['Sex']].values
```

```python
def intoBinary(x):
    for i in range(len(x)):
        if x[i] == 'female':
            x[i] = 0
        elif x[i] == 'male':
            x[i] = 1
    return x

# Example usage
x1 = intoBinary(x1)
print(x1)
```

```python
X=np.column_stack((X, x1))
```

```python
print(X)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)


from sklearn.preprocessing import Normalizer

#scaler= Normalizer().fit(X) # the scaler is fitted to the  dataset
#X= scaler.transform(X) # the scaler is applied to the dataset
 
```

```python
print(X)
```

```python
dataset.plot.scatter(x='Fare',y='Age', c='Survived', colormap='winter')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X.shape
#y.shape
```

```python
log = LogisticRegression()
log.fit(X, y)
```

```python
y_pred = log.predict(X)
```

```python
y_pred
```

```python
n = 200
plt.plot(y[:n])
plt.plot(y_pred[:n], 'o')
plt.show()
```

```python
y_pred[:n]
```

```python
s = log.score(X, y)
s
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, y_pred)
print(cm)
```

```python
b = log.intercept_[0]
w1 = log.coef_[0, 0]
w2 = log.coef_[0, 1]

x1 = X[:, 0]
x2 = -(b+w1*x1)/w2
x2[:10]
```

```python
x_pos = np.array([[x[4], x[7]] for x in dataset.values if x[0] >= 0.5]) 
x_neg = np.array([[x[4], x[7]] for x in dataset.values if x[0] <= 0.5])
```

```python
plt.plot(x_pos[:, 0], x_pos[:, 1], '+')
plt.plot(x_neg[:, 0], x_neg[:, 1], '*')
plt.plot(x1, x2)                                                                                                                        
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Plot confusion matrix
import seaborn as sns
import pandas as pd
# confusion matrix sns heatmap 
## https://www.kaggle.com/agungor2/various-confusion-matrix-plots
ax = plt.axes() 
df_cm = cm
sns.heatmap(df_cm, annot=True, annot_kws={"size": 30}, fmt='d',cmap="Blues", ax = ax )
ax.set_title('Confusion Matrix')
plt.show()
```
